```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [35, 10, 45, 10]
colors = ['#FF5733', '#33C1FF', '#28B463', '#AF7AC5']  # Hexadecimal colors for the slices
explode = (0.1, 0, 0, 0)  # Explode the 2nd slice (B)

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, colors=colors)

ax.axis('equal')

ax.legend(labels, title="Categories", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Pie Chart with Exploded Slice and Legend')
plt.show()
```
